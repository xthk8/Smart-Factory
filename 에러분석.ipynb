```python
import pandas as pd

total_date = pd.read_csv("C:/Users/USER/Desktop/설비학회/Dataset_장비이상 조기탐지 AI 데이터셋/data/5공정_180sec/total_date.csv")
error_lot_list = pd.read_csv("C:/Users/USER/Desktop/설비학회/Error Lot list.csv")
```

```python

# Correct the column names
error_lot_list.columns = ['oper_date', 'bd1', 'bd2', 'bd3', 'bd4', 'bd5', 'bd6', 'bd7', 'bd8', 'bd9', 'bd10', 'bd11']
total_date.columns = ['Index', 'process', 'Time', 'temp', 'current', 'date']

# Merging data
merged_data = pd.merge(
    left=error_lot_list.melt(id_vars=['oper_date'], value_vars=['bd1', 'bd2', 'bd3', 'bd4', 'bd5', 'bd6', 'bd7', 'bd8', 'bd9', 'bd10', 'bd11'], value_name='process'),
    right=total_date.groupby(['date', 'process'])[['temp', 'current']].mean().reset_index(),
    left_on=['oper_date', 'process'],
    right_on=['date', 'process'],
    how='right'
)

# Adding a column to check if the process matches or not
merged_data['condition_met'] = merged_data['oper_date'].notna()

# Removing 'oper_date' and 'variable' columns
final_data = merged_data.drop(columns=['oper_date', 'variable'])

# Converting 'process' to integer
final_data['process'] = final_data['process'].astype(int)
final_data.rename(columns={'temp': 'temp_avg', 'current': 'current_avg'}, inplace=True)

# Displaying the first few rows of the final data
final_data
final_data.to_csv("C:/Users/USER/Desktop/설비학회/Dataset_장비이상 조기탐지 AI 데이터셋/data/5공정_180sec/total_analysis.csv")
```
